# Mean Squared Displacement (MSD) analysis 

In this tutorial we are going to learn how to use the post_processing module to compute the **Mean Squared Displacement** (MSD) from two-dimensional molecular dynamics simulation snapshots. First of all, make sure to have compiled the module to this directory by doing on you're bash: 

- f2py3 -c ../../../lib/post_processing.f90 -m post_processing

Done that, we are ready to start our analysis. In this directory, you will find the simulation snapshots in file "snap_LAMMPS_2d.xyz". And the LAMMPS computed MSD in file "lammps_msd.dat" to compare with our data generated from the snapshot MSD module interpreter **post_processing**. 

## Starting up 

We need to add some useful modules, henceforth numpy, post_processing, matplotlib will be imported by: 

```python
from post_processing import * 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

And then, we proceed by computing the msd and time arrays from post_processing by letting it know some parameters:

- Number of particle: $N = 200$ 
- Snapshot file: "snap_LAMMPS_2d.xyz" 
- Number of snapshots: 1000
- Box side length: L = 20 
- And the number of snaps times the simulation dt: $nsnaps * dt = 10$

Knowing that, one can compute the msd by the bellow command: 

```python
msd, t = msd_evaluation(200,"snap_LAMMPS_2d.xyz",1000,20,10)
```

Now we will read in LAMMPS computed MSD to compare with our generated MSD from the snapshots:

```python
data=np.loadtxt("lammps_msd.dat")
```

To plot our data and LAMMPS computed MSD: 

```python
plt.plot(t,msd,ls='-', label="from snapshot")
plt.plot(data[:,0]*0.01,data[:,1], ls ='-', label="from LAMMPS")

plt.ylabel(r"MSD")
plt.xlabel(r"$t$")
plt.legend(loc="best")
```

From plot we are sure that our post_processing module computes the same MSD, which is computed directly by LAMMPS package.
